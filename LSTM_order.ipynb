```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv('traffic.csv')

# Inspect the first few rows to verify column names and structure
print(df.head())
print("Data shape:", df.shape)
```

```python
import numpy as np

# Load the indices from the saved text files
train_idx = np.loadtxt("train_order_idx.txt", dtype=int)
val_idx   = np.loadtxt("val_order_idx.txt", dtype=int)
test_idx  = np.loadtxt("test_order_idx.txt", dtype=int)

# Re-create the training, validation, and test DataFrames from df_junction1
train_data = df_junction1.iloc[train_idx].reset_index(drop=True)
val_data   = df_junction1.iloc[val_idx].reset_index(drop=True)
test_data  = df_junction1.iloc[test_idx].reset_index(drop=True)

print("Train, Val, and Test sets created:")
print(" - Train data shape:", train_data.shape)
print(" - Validation data shape:", val_data.shape)
print(" - Test data shape:", test_data.shape)
```

```python
# ... [Keep imports and initial data loading/splitting cells 58b7e3fd, 365563bc, 42cf2ac2, 8de565d7] ...

# === CELL 3a153ab2 (Modified for LSTM with Lags) ===
# GPU setup
import torch
from torch import nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
import matplotlib.pyplot as plt
import numpy as np
import time
import pandas as pd
from sklearn.preprocessing import StandardScaler, QuantileTransformer
import xgboost as xgb # Import XGBoost
from sklearn.metrics import mean_squared_error, mean_absolute_error # Import metrics

print("Torch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())

if torch.cuda.is_available():
    print("CUDA version:", torch.version.cuda)
    print("Device name:", torch.cuda.get_device_name(0))
else:
    print("CUDA is NOT available. Using CPU.")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# === 1. Feature Engineering (Lags and Temporal) ===
print("Extracting lag and temporal features...")

def create_all_features(df, target_col='Vehicles', max_lag=168):
    df_feat = df.copy()
    df_feat['DateTime'] = pd.to_datetime(df_feat['DateTime'])

    # --- Lag Features ---
    print(f"Creating lag features up to {max_lag} hours...")
    for lag in range(1, max_lag + 1):
        df_feat[f'lag_{lag}'] = df_feat[target_col].shift(lag)

    # --- Temporal Features ---
    df_feat['hour'] = df_feat['DateTime'].dt.hour
    df_feat['day_of_week'] = df_feat['DateTime'].dt.dayofweek
    df_feat['day'] = df_feat['DateTime'].dt.day
    df_feat['month'] = df_feat['DateTime'].dt.month
    # Cyclical encoding
    df_feat['hour_sin'] = np.sin(2 * np.pi * df_feat['hour']/24)
    df_feat['hour_cos'] = np.cos(2 * np.pi * df_feat['hour']/24)
    df_feat['day_of_week_sin'] = np.sin(2 * np.pi * df_feat['day_of_week']/7)
    df_feat['day_of_week_cos'] = np.cos(2 * np.pi * df_feat['day_of_week']/7)
    df_feat['month_sin'] = np.sin(2 * np.pi * df_feat['month']/12)
    df_feat['month_cos'] = np.cos(2 * np.pi * df_feat['month']/12)
    # Indicators
    df_feat['is_weekend'] = df_feat['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)
    df_feat['is_workday'] = (~df_feat['is_weekend']).astype(int)
    df_feat['morning_rush'] = ((df_feat['hour'] >= 7) & (df_feat['hour'] <= 9)).astype(int)
    df_feat['evening_rush'] = ((df_feat['hour'] >= 16) & (df_feat['hour'] <= 19)).astype(int)
    df_feat['is_rush_hour'] = ((df_feat['morning_rush'] | df_feat['evening_rush']) & df_feat['is_workday']).astype(int)

    # Drop rows with NaNs created by lags
    df_feat.dropna(inplace=True)
    return df_feat

# Define feature sets
max_lag = 168
lag_features = [f'lag_{i}' for i in range(1, max_lag + 1)]
temporal_features = [ # Keep these separate for the refinement stage
    'hour_sin', 'hour_cos', 'day_of_week_sin', 'day_of_week_cos',
    'month_sin', 'month_cos', 'is_weekend', 'day',
    'morning_rush', 'evening_rush', 'is_rush_hour', 'is_workday'
]
target_col = 'Vehicles'

# Apply feature engineering to train, val, test
# Important: Apply to the original splits to avoid data leakage across sets
train_data_feat = create_all_features(train_data.copy(), target_col=target_col, max_lag=max_lag)
val_data_feat = create_all_features(val_data.copy(), target_col=target_col, max_lag=max_lag)
test_data_feat = create_all_features(test_data.copy(), target_col=target_col, max_lag=max_lag)

print(f"Shapes after feature engineering & NaN drop:")
print(f"Train: {train_data_feat.shape}, Val: {val_data_feat.shape}, Test: {test_data_feat.shape}")

# === 2. Scaling ===
print("Scaling features...")

# Scale lag features (LSTM input)
lag_scaler = StandardScaler() # StandardScaler often works well for LSTM inputs
X_train_lags = lag_scaler.fit_transform(train_data_feat[lag_features])
X_val_lags = lag_scaler.transform(val_data_feat[lag_features])
X_test_lags = lag_scaler.transform(test_data_feat[lag_features])

# Scale temporal features (Refinement input)
temporal_scaler = StandardScaler()
train_temporal = temporal_scaler.fit_transform(train_data_feat[temporal_features])
val_temporal = temporal_scaler.transform(val_data_feat[temporal_features])
test_temporal = temporal_scaler.transform(test_data_feat[temporal_features])

# Scale target variable (using QuantileTransformer as before)
target_scaler = QuantileTransformer(output_distribution='normal', n_quantiles=min(1000, len(train_data_feat)))
y_train_scaled = target_scaler.fit_transform(train_data_feat[[target_col]]).flatten()
y_val_scaled = target_scaler.transform(val_data_feat[[target_col]]).flatten()
y_test_scaled = target_scaler.transform(test_data_feat[[target_col]]).flatten()

# Keep original targets for final evaluation
y_train_orig = train_data_feat[target_col].values
y_val_orig = val_data_feat[target_col].values
y_test_orig = test_data_feat[target_col].values

# === 3. Prepare Data for LSTM (Lags as sequence) ===
# The LSTM expects input shape: (batch_size, seq_length, num_features)
# Here, seq_length is 1 (we predict the next step based on the 168 lags),
# and num_features is 168 (the number of lags).
# OR, we can treat the 168 lags as the sequence itself (seq_length=168, num_features=1).
# Let's try the second approach as it's more common for LSTMs.

seq_length = max_lag # Treat the lags as the sequence
num_features = 1     # Each lag is one feature at a time step

X_train_lstm = X_train_lags.reshape(-1, seq_length, num_features)
X_val_lstm = X_val_lags.reshape(-1, seq_length, num_features)
X_test_lstm = X_test_lags.reshape(-1, seq_length, num_features)

# Convert to Tensors
X_train_lstm_tensor = torch.tensor(X_train_lstm, dtype=torch.float32).to(device)
X_val_lstm_tensor = torch.tensor(X_val_lstm, dtype=torch.float32).to(device)
X_test_lstm_tensor = torch.tensor(X_test_lstm, dtype=torch.float32).to(device)

y_train_tensor = torch.tensor(y_train_scaled, dtype=torch.float32).unsqueeze(1).to(device) # Add feature dim
y_val_tensor = torch.tensor(y_val_scaled, dtype=torch.float32).unsqueeze(1).to(device)
y_test_tensor = torch.tensor(y_test_scaled, dtype=torch.float32).unsqueeze(1).to(device)

# Temporal features tensors (will be used in refinement stage)
train_temporal_tensor = torch.tensor(train_temporal, dtype=torch.float32).to(device)
val_temporal_tensor = torch.tensor(val_temporal, dtype=torch.float32).to(device)
test_temporal_tensor = torch.tensor(test_temporal, dtype=torch.float32).to(device)

print(f"LSTM Input Shape (Train): {X_train_lstm_tensor.shape}")
print(f"LSTM Target Shape (Train): {y_train_tensor.shape}")
print(f"Temporal Features Shape (Train): {train_temporal_tensor.shape}")

# === 4. Define LSTM Model (Simpler version for lag input) ===
class LagLSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim=1, num_layers=2, dropout=0.3, attention_dim=64):
        """
        Enhanced LSTM model with Bidirectionality and Attention.

        Args:
            input_dim (int): Number of input features (1 for lag value).
            hidden_dim (int): Number of features in the hidden state h.
            output_dim (int): Number of output features (1 for prediction).
            num_layers (int): Number of recurrent layers.
            dropout (float): Dropout probability.
            attention_dim (int): Dimension of the attention mechanism's internal layer.
        """
        super(LagLSTMModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.num_layers = num_layers
        self.bidirectional = True # Make LSTM bidirectional
        # LSTM output dimension will be hidden_dim * 2 because it's bidirectional
        lstm_output_dim = hidden_dim * 2 if self.bidirectional else hidden_dim

        self.lstm = nn.LSTM(
            input_dim,
            hidden_dim,
            num_layers=num_layers,
            batch_first=True, # Input: (batch, seq_len, features)
            dropout=dropout if num_layers > 1 else 0,
            bidirectional=self.bidirectional
        )

        # --- Attention Mechanism ---
        # Fully connected layers to compute attention scores
        self.attention_fc1 = nn.Linear(lstm_output_dim, attention_dim)
        self.attention_fc2 = nn.Linear(attention_dim, 1)
        # --- End Attention Mechanism ---

        # Layer Normalization - applied after LSTM and after Attention context
        self.layer_norm_lstm = nn.LayerNorm(lstm_output_dim)
        self.layer_norm_context = nn.LayerNorm(lstm_output_dim)

        # Output layer takes the context vector from attention
        self.fc = nn.Linear(lstm_output_dim, output_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        # x shape: (batch_size, seq_length, input_dim)

        # LSTM layer
        # lstm_out shape: (batch_size, seq_length, hidden_dim * num_directions)
        # hn shape: (num_layers * num_directions, batch_size, hidden_dim)
        # cn shape: (num_layers * num_directions, batch_size, hidden_dim)
        lstm_out, (hn, cn) = self.lstm(x)

        # Apply Layer Normalization to the output of the LSTM
        lstm_out_norm = self.layer_norm_lstm(lstm_out)

        # --- Attention Calculation ---
        # Pass LSTM output through the first attention layer and apply tanh activation
        # attn_intermed shape: (batch_size, seq_length, attention_dim)
        attn_intermed = torch.tanh(self.attention_fc1(lstm_out_norm))
        # Pass through the second attention layer to get scores
        # attn_scores shape: (batch_size, seq_length, 1)
        attn_scores = self.attention_fc2(attn_intermed)
        # Apply softmax to get attention weights (normalized scores)
        # attn_weights shape: (batch_size, seq_length, 1)
        attn_weights = F.softmax(attn_scores, dim=1)

        # Calculate the context vector by taking a weighted sum of LSTM outputs
        # context shape: (batch_size, hidden_dim * num_directions)
        context = torch.sum(attn_weights * lstm_out_norm, dim=1)
        # --- End Attention Calculation ---

        # Apply Layer Normalization to the context vector
        context_norm = self.layer_norm_context(context)

        # Apply dropout
        out = self.dropout(context_norm)
        # Final fully connected layer
        out = self.fc(out)
        return out

# Instantiate LSTM model
lstm_input_dim = num_features # 1
lstm_hidden_dim = 256 # Can be tuned
lstm_output_dim = 1
lstm_num_layers = 2
lstm_dropout = 0.05
lstm_attention_dim = 64 # Dimension for attention mechanism
lstm_model = LagLSTMModel(
    input_dim=lstm_input_dim,
    hidden_dim=lstm_hidden_dim,
    output_dim=lstm_output_dim,
    num_layers=lstm_num_layers,
    dropout=lstm_dropout,
    attention_dim=lstm_attention_dim
).to(device)

print("\nLSTM Model Architecture:")
print(lstm_model)

# === 5. Train LSTM Model ===
print("\n--- Training LSTM Model ---")
lstm_criterion = nn.MSELoss() # Standard MSE for the LSTM stage
lstm_optimizer = torch.optim.AdamW(lstm_model.parameters(), lr=0.001, weight_decay=1e-5)

lstm_train_dataset = TensorDataset(X_train_lstm_tensor, y_train_tensor)
lstm_train_loader = DataLoader(lstm_train_dataset, batch_size=64, shuffle=True)
# --- Validation DataLoader ---
lstm_val_dataset = TensorDataset(X_val_lstm_tensor, y_val_tensor)
# Use a larger batch size for validation if possible, as no gradients are computed
lstm_val_loader = DataLoader(lstm_val_dataset, batch_size=64, shuffle=False)
lstm_epochs = 200 # Adjust as needed
lstm_patience = 40
lstm_patience_counter = 0
best_lstm_val_loss = float('inf')
lstm_train_losses = []
lstm_val_losses = []

for epoch in range(lstm_epochs):
    lstm_model.train()
    epoch_loss = 0
    for batch_X, batch_y in lstm_train_loader:
        lstm_optimizer.zero_grad()
        outputs = lstm_model(batch_X)
        loss = lstm_criterion(outputs, batch_y)
        loss.backward()
        lstm_optimizer.step()
        epoch_loss += loss.item()

    avg_train_loss = epoch_loss / len(lstm_train_loader)
    lstm_train_losses.append(avg_train_loss)

    # Validation
    lstm_model.eval()
    epoch_val_loss = 0
    with torch.no_grad():
        for batch_X_val, batch_y_val in lstm_val_loader: # Iterate through validation batches
            val_outputs = lstm_model(batch_X_val)
            loss = lstm_criterion(val_outputs, batch_y_val)
            epoch_val_loss += loss.item()

    avg_val_loss = epoch_val_loss / len(lstm_val_loader) # Calculate average validation loss
    lstm_val_losses.append(avg_val_loss)
    # --- End Validation (Batched) ---

    print(f"Epoch {epoch+1}/{lstm_epochs}, LSTM Train Loss: {avg_train_loss:.6f}, LSTM Val Loss: {avg_val_loss:.6f}") # Use avg_val_loss

    # Early stopping (use avg_val_loss)
    if avg_val_loss < best_lstm_val_loss:
        best_lstm_val_loss = avg_val_loss
        torch.save(lstm_model.state_dict(), "best_lag_lstm_model.pt")
        lstm_patience_counter = 0
        print(f"  -> New best validation loss: {best_lstm_val_loss:.6f}. Saving model.") # Added print statement
    else:
        lstm_patience_counter += 1
        if lstm_patience_counter >= lstm_patience:
            print(f"Early stopping LSTM training at epoch {epoch+1}")
            break

# Load best LSTM model
import os
if os.path.exists("best_lag_lstm_model.pt"):
    lstm_model.load_state_dict(torch.load("best_lag_lstm_model.pt"))
    print("Best LSTM model loaded.")
else:
    print("Warning: 'best_lag_lstm_model.pt' not found. Using the model from the last epoch.")

# Plot LSTM losses
plt.figure(figsize=(10, 4))
plt.plot(lstm_train_losses, label='LSTM Train Loss')
plt.plot(lstm_val_losses, label='LSTM Val Loss')
plt.title('LSTM Training History')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# === 6. Generate LSTM Predictions (Intermediate Step) ===
print("\nGenerating intermediate LSTM predictions (batched)...")

def predict_in_batches(model, X_tensor, batch_size=64):
    """Generates predictions in batches to avoid OOM errors."""
    model.eval()
    device = next(model.parameters()).device
    all_preds_scaled = []
    dataset = TensorDataset(X_tensor)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

    with torch.no_grad():
        for (batch_X,) in dataloader: # Unpack the tuple from DataLoader
            batch_X = batch_X.to(device)
            outputs = model(batch_X).squeeze()
            all_preds_scaled.append(outputs.cpu().numpy())

    # Concatenate results from all batches
    predictions_scaled = np.concatenate(all_preds_scaled)
    return predictions_scaled.flatten() # Ensure 1D array

# Generate predictions using the batch function
lstm_train_pred_scaled = predict_in_batches(lstm_model, X_train_lstm_tensor)
lstm_val_pred_scaled   = predict_in_batches(lstm_model, X_val_lstm_tensor)
lstm_test_pred_scaled  = predict_in_batches(lstm_model, X_test_lstm_tensor)

# === 7. Prepare Data for Refinement Stage ===
# Features: LSTM prediction + Temporal Features
# Target: Actual Scaled Vehicle Count

# Add LSTM prediction as a new feature (needs reshaping)
X_train_refine = np.concatenate((lstm_train_pred_scaled.reshape(-1, 1), train_temporal), axis=1)
X_val_refine = np.concatenate((lstm_val_pred_scaled.reshape(-1, 1), val_temporal), axis=1)
X_test_refine = np.concatenate((lstm_test_pred_scaled.reshape(-1, 1), test_temporal), axis=1)

# Targets remain the same scaled targets used for LSTM
y_train_refine = y_train_scaled
y_val_refine = y_val_scaled
y_test_refine = y_test_scaled

print(f"\nRefinement Input Shape (Train): {X_train_refine.shape}")
print(f"Refinement Target Shape (Train): {y_train_refine.shape}")

```

```python
# === 9. Train and Evaluate Refinement Model (Simple NN) ===
print("\n--- Training Refinement Model (Enhanced NN) ---") # Modified print

import torch.nn.init as init

class RefinementNN(nn.Module):
    def __init__(self, input_dim, hidden_dims=[128, 64, 32], output_dim=1, dropout_rate=0.4, activation_fn=nn.LeakyReLU(negative_slope=0.01)):
        """
        Enhanced Feed-Forward Neural Network for refinement with residual connections,
        configurable layers/activation, and weight initialization.

        Args:
            input_dim (int): Dimension of the input features.
            hidden_dims (list): List of integers specifying the size of each hidden layer.
            output_dim (int): Dimension of the output (usually 1).
            dropout_rate (float): Dropout probability.
            activation_fn (nn.Module): PyTorch activation function module instance.
        """
        super(RefinementNN, self).__init__()
        self.layers = nn.ModuleList()
        self.batch_norms = nn.ModuleList()
        self.residual_projections = nn.ModuleList() # For matching dimensions in residual connections
        self.dropout = nn.Dropout(dropout_rate)
        self.activation_fn = activation_fn

        current_dim = input_dim
        for i, h_dim in enumerate(hidden_dims):
            linear_layer = nn.Linear(current_dim, h_dim)
            self._initialize_weights(linear_layer) # Initialize weights
            self.layers.append(linear_layer)
            self.batch_norms.append(nn.BatchNorm1d(h_dim))

            # Add projection layer for residual connection if dimensions don't match
            if current_dim != h_dim:
                projection = nn.Linear(current_dim, h_dim)
                self._initialize_weights(projection)
                self.residual_projections.append(projection)
            else:
                self.residual_projections.append(nn.Identity()) # No projection needed

            current_dim = h_dim

        # Output layer
        self.output_layer = nn.Linear(current_dim, output_dim)
        self._initialize_weights(self.output_layer) # Initialize output layer weights

    def _initialize_weights(self, layer):
        # Use Kaiming (He) initialization for linear layers, suitable for ReLU-like activations
        if isinstance(layer, nn.Linear):
            init.kaiming_uniform_(layer.weight, a=0.01 if isinstance(self.activation_fn, nn.LeakyReLU) else 0, mode='fan_in', nonlinearity='leaky_relu' if isinstance(self.activation_fn, nn.LeakyReLU) else 'relu')
            if layer.bias is not None:
                init.constant_(layer.bias, 0)

    def forward(self, x):
        residual = x # Store initial input for the first residual connection

        # Pass through hidden layers
        for i, layer in enumerate(self.layers):
            identity = self.residual_projections[i](residual) # Project residual if needed

            x = layer(x)
            # Apply BatchNorm only if the batch size is greater than 1
            if x.size(0) > 1:
                 x = self.batch_norms[i](x)

            x = self.activation_fn(x)
            x = x + identity # Add residual connection *after* activation
            x = self.dropout(x) # Apply dropout after residual connection

            residual = x # Update residual for the next block

        # Output layer
        x = self.output_layer(x)
        return x

# Prepare data for NN refinement (Using original X_train_refine etc. which don't have lagged residuals)
X_train_refine_tensor = torch.tensor(X_train_refine_df[base_feature_names + interaction_features].values, dtype=torch.float32).to(device) # Use updated features
X_val_refine_tensor = torch.tensor(X_val_refine_df[base_feature_names + interaction_features].values, dtype=torch.float32).to(device)
X_test_refine_tensor = torch.tensor(X_test_refine_df[base_feature_names + interaction_features].values, dtype=torch.float32).to(device)

# Target tensors need to be reshaped for NN loss (Using original y_train_refine etc.)
y_train_refine_tensor = torch.tensor(y_train_refine, dtype=torch.float32).unsqueeze(1).to(device)
y_val_refine_tensor = torch.tensor(y_val_refine, dtype=torch.float32).unsqueeze(1).to(device)
y_test_refine_tensor = torch.tensor(y_test_refine, dtype=torch.float32).unsqueeze(1).to(device)

# --- Instantiate the Enhanced NN Model ---
# You can change the activation function here, e.g., activation_fn=nn.ReLU() or nn.GELU()
nn_refine_model = RefinementNN(
    input_dim=X_train_refine_tensor.shape[1],
    hidden_dims=[128, 128, 64, 64], # Example: Slightly deeper architecture
    dropout_rate=0.3,             # Example: Adjusted dropout
    activation_fn=nn.LeakyReLU(negative_slope=0.01) # Keep LeakyReLU or change
).to(device)

print("\nEnhanced NN Refinement Model Architecture:")
print(nn_refine_model)

nn_criterion = nn.MSELoss()
nn_optimizer = torch.optim.AdamW(nn_refine_model.parameters(), lr=0.001, weight_decay=1e-4)
# --- Optional: Add a Learning Rate Scheduler ---
# from torch.optim.lr_scheduler import ReduceLROnPlateau
# nn_scheduler = ReduceLROnPlateau(nn_optimizer, 'min', factor=0.1, patience=5, verbose=True)

nn_train_dataset = TensorDataset(X_train_refine_tensor, y_train_refine_tensor)
nn_train_loader = DataLoader(nn_train_dataset, batch_size=128, shuffle=True)

nn_epochs = 150 # Increased epochs slightly, adjust as needed
nn_patience = 20 # Increased patience slightly
nn_patience_counter = 0
best_nn_val_loss = float('inf')
nn_train_losses = []
nn_val_losses = []

for epoch in range(nn_epochs):
    nn_refine_model.train()
    epoch_loss = 0
    for batch_X, batch_y in nn_train_loader:
        nn_optimizer.zero_grad()
        outputs = nn_refine_model(batch_X)
        loss = nn_criterion(outputs, batch_y)
        loss.backward()
        # Optional: Gradient Clipping
        # torch.nn.utils.clip_grad_norm_(nn_refine_model.parameters(), max_norm=1.0)
        nn_optimizer.step()
        epoch_loss += loss.item()

    avg_train_loss = epoch_loss / len(nn_train_loader)
    nn_train_losses.append(avg_train_loss)

    # Validation
    nn_refine_model.eval()
    with torch.no_grad():
        val_outputs = nn_refine_model(X_val_refine_tensor)
        val_loss = nn_criterion(val_outputs, y_val_refine_tensor).item()
        nn_val_losses.append(val_loss)

    # --- Optional: Step the scheduler based on validation loss ---
    # nn_scheduler.step(val_loss)

    print(f"Epoch {epoch+1}/{nn_epochs}, NN Train Loss: {avg_train_loss:.6f}, NN Val Loss: {val_loss:.6f}")

    # Early stopping
    if val_loss < best_nn_val_loss:
        best_nn_val_loss = val_loss
        torch.save(nn_refine_model.state_dict(), "best_nn_refine_model.pt")
        nn_patience_counter = 0
        print(f"  -> New best validation loss: {best_nn_val_loss:.6f}. Saving model.") # Added print statement
    else:
        nn_patience_counter += 1
        if nn_patience_counter >= nn_patience:
            print(f"Early stopping NN refinement training at epoch {epoch+1}")
            break

# Load best NN model
import os # Ensure os is imported
if os.path.exists("best_nn_refine_model.pt"):
    nn_refine_model.load_state_dict(torch.load("best_nn_refine_model.pt"))
    print("Best NN refinement model loaded.")
else:
    print("Warning: 'best_nn_refine_model.pt' not found. Using the model from the last epoch.")


# Plot NN losses
plt.figure(figsize=(10, 4))
plt.plot(nn_train_losses, label='NN Refine Train Loss')
plt.plot(nn_val_losses, label='NN Refine Val Loss')
plt.title('NN Refinement Training History')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# Predict with NN
nn_refine_model.eval()
with torch.no_grad():
    # Use predict_in_batches for potentially large test sets
    nn_train_pred_scaled = predict_in_batches(nn_refine_model, X_train_refine_tensor)
    nn_val_pred_scaled = predict_in_batches(nn_refine_model, X_val_refine_tensor)
    nn_test_pred_scaled = predict_in_batches(nn_refine_model, X_test_refine_tensor)
    # nn_train_pred_scaled = nn_refine_model(X_train_refine_tensor).squeeze().cpu().numpy() # Original way
    # nn_val_pred_scaled = nn_refine_model(X_val_refine_tensor).squeeze().cpu().numpy()     # Original way
    # nn_test_pred_scaled = nn_refine_model(X_test_refine_tensor).squeeze().cpu().numpy()    # Original way


# Inverse transform final predictions
final_train_pred_nn = target_scaler.inverse_transform(nn_train_pred_scaled.reshape(-1, 1)).flatten()
final_val_pred_nn = target_scaler.inverse_transform(nn_val_pred_scaled.reshape(-1, 1)).flatten()
final_test_pred_nn = target_scaler.inverse_transform(nn_test_pred_scaled.reshape(-1, 1)).flatten()

# Clip predictions (important after inverse transform)
final_train_pred_nn = np.clip(final_train_pred_nn, 0, None)
final_val_pred_nn = np.clip(final_val_pred_nn, 0, None)
final_test_pred_nn = np.clip(final_test_pred_nn, 0, None)


# Evaluate NN Refinement
print("\n--- NN Refinement Evaluation (Test Set) ---")
# Use original y_test_orig_final for evaluation (ensure it exists and aligns)
if 'y_test_orig_final' in locals() and len(y_test_orig_final) == len(final_test_pred_nn):
    test_rmse_nn = np.sqrt(mean_squared_error(y_test_orig_final, final_test_pred_nn))
    test_mae_nn = mean_absolute_error(y_test_orig_final, final_test_pred_nn)
    print(f"NN Refined - Test RMSE: {test_rmse_nn:.2f}, Test MAE: {test_mae_nn:.2f}")
else:
    print("Error: Cannot evaluate NN refinement. 'y_test_orig_final' missing or length mismatch.")
    test_rmse_nn, test_mae_nn = float('inf'), float('inf') # Set to infinity if evaluation fails

# === 10. Final Comparison Plot ===
print("\n--- Final Comparison on Test Set ---")

# Get LSTM-only predictions (inverse transformed) - Use original ones
lstm_test_pred_orig = target_scaler.inverse_transform(lstm_test_pred_scaled_final.reshape(-1, 1)).flatten()
lstm_test_rmse = np.sqrt(mean_squared_error(y_test_orig_final, lstm_test_pred_orig))
lstm_test_mae = mean_absolute_error(y_test_orig_final, lstm_test_pred_orig)
print(f"LSTM Only - Test RMSE: {lstm_test_rmse:.2f}, Test MAE: {lstm_test_mae:.2f}")


plt.figure(figsize=(15, 7))
# Plot a subset for clarity (e.g., first 200 points of the test set)
plot_range = range(min(200, len(y_test_orig_final))) # Use y_test_orig_final length
plt.plot(plot_range, y_test_orig_final[plot_range], label='Actual', marker='.', linestyle='-', alpha=0.7)
plt.plot(plot_range, lstm_test_pred_orig[plot_range], label=f'LSTM Only (RMSE: {lstm_test_rmse:.2f})', marker='^', linestyle=':', alpha=0.8)
# Ensure final_test_pred_xgb has the same length as y_test_orig_final before plotting
if len(final_test_pred_xgb) == len(y_test_orig_final):
    plt.plot(plot_range, final_test_pred_xgb[plot_range], label=f'LSTM + XGBoost (RMSE: {test_rmse_xgb:.2f})', marker='x', linestyle='--', alpha=0.9)
else:
    print(f"Warning: XGBoost prediction length ({len(final_test_pred_xgb)}) doesn't match target length ({len(y_test_orig_final)}). Skipping plot.")
# Ensure final_test_pred_nn has the same length as y_test_orig_final before plotting
if len(final_test_pred_nn) == len(y_test_orig_final):
    plt.plot(plot_range, final_test_pred_nn[plot_range], label=f'LSTM + NN (RMSE: {test_rmse_nn:.2f})', marker='s', linestyle='-.', alpha=0.9)
else:
    print(f"Warning: NN prediction length ({len(final_test_pred_nn)}) doesn't match target length ({len(y_test_orig_final)}). Skipping plot.")


plt.title('Comparison of Models on Test Set')
plt.xlabel('Time Step')
plt.ylabel('Vehicle Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Final Results Summary
print("\n===== FINAL RESULTS SUMMARY (Test Set) =====")
results_data = {
    'Model': ['LSTM Only', 'LSTM + XGBoost', 'LSTM + NN'],
    'RMSE': [lstm_test_rmse, test_rmse_xgb, test_mae_nn], # Use test_mae_nn for NN MAE
    'MAE': [lstm_test_mae, test_mae_xgb, test_mae_nn]
}
results_df = pd.DataFrame(results_data)
print(results_df.sort_values('RMSE'))
```

```python
# === 11. Train LSTM to Mimic LSTM+XGBoost (Knowledge Distillation - Enhanced with Separate Model) ===
print("\n--- Training LSTM to Mimic LSTM+XGBoost (Enhanced Distillation with Separate Model) ---")

# --- Import Scheduler ---
from torch.optim.lr_scheduler import ReduceLROnPlateau
import os
import matplotlib.pyplot as plt
import torch # Ensure torch, nn, F are available
from torch import nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset
import numpy as np # Ensure numpy is available

# Check if XGBoost results exist
if 'final_train_pred_scaled' not in locals() or 'final_val_pred_scaled' not in locals():
    print("Error: LSTM+XGBoost predictions ('final_train_pred_scaled', 'final_val_pred_scaled') not found. Run cell 05cc2ca2 first.")
# Check if temporal tensors exist
elif 'train_temporal_tensor' not in locals() or 'val_temporal_tensor' not in locals() or 'test_temporal_tensor' not in locals():
    print("Error: Scaled temporal feature tensors ('train_temporal_tensor', etc.) not found. Ensure cell 663da4ae ran successfully.")
# Check if ORIGINAL scaled targets exist
elif 'y_train_scaled' not in locals() or 'y_val_scaled' not in locals():
     print("Error: Original scaled targets ('y_train_scaled', 'y_val_scaled') not found. Ensure cell 663da4ae ran successfully.")
else:
    # --- Prepare Targets ---
    # 1. Teacher Targets (Scaled LSTM+XGBoost Predictions)
    y_train_teacher_target_np = np.array(final_train_pred_scaled)
    y_val_teacher_target_np = np.array(final_val_pred_scaled)
    y_train_teacher_target = torch.tensor(y_train_teacher_target_np, dtype=torch.float32).unsqueeze(1).to(device)
    y_val_teacher_target = torch.tensor(y_val_teacher_target_np, dtype=torch.float32).unsqueeze(1).to(device)

    # 2. Ground Truth Targets (Original Scaled Vehicle Counts)
    y_train_gt_target_np = np.array(y_train_scaled)
    y_val_gt_target_np = np.array(y_val_scaled)
    y_train_gt_target = torch.tensor(y_train_gt_target_np, dtype=torch.float32).unsqueeze(1).to(device)
    y_val_gt_target = torch.tensor(y_val_gt_target_np, dtype=torch.float32).unsqueeze(1).to(device)

    print(f"Teacher Target Shape (Train): {y_train_teacher_target.shape}")
    print(f"Ground Truth Target Shape (Train): {y_train_gt_target.shape}")

    # --- Prepare Input Features ---
    num_temporal_features = train_temporal_tensor.shape[1]
    print(f"Lag Input Shape for Distillation (Train): {X_train_lstm_tensor.shape}")
    print(f"Temporal Input Shape for Distillation (Train): {train_temporal_tensor.shape}")

    # === Define NEW LSTM Model Class Specifically for Distillation ===
    class DistillLagLSTMModel(nn.Module):
        # --- This class definition is specific to this cell ---
        def __init__(self, input_dim, hidden_dim, output_dim=1, num_layers=2, dropout=0.3, attention_dim=64, temporal_dim=0):
            """ LSTM model designed for distillation, accepting temporal features. """
            super(DistillLagLSTMModel, self).__init__()
            self.hidden_dim = hidden_dim
            self.num_layers = num_layers
            self.bidirectional = True
            lstm_output_dim = hidden_dim * 2 if self.bidirectional else hidden_dim
            self.temporal_dim = temporal_dim # Store temporal dimension

            self.lstm = nn.LSTM(
                input_dim, hidden_dim, num_layers=num_layers, batch_first=True,
                dropout=dropout if num_layers > 1 else 0, bidirectional=self.bidirectional
            )
            self.attention_fc1 = nn.Linear(lstm_output_dim, attention_dim)
            self.attention_fc2 = nn.Linear(attention_dim, 1)
            self.layer_norm_lstm = nn.LayerNorm(lstm_output_dim)
            # Adjust context norm input dim based on temporal_dim
            self.layer_norm_context = nn.LayerNorm(lstm_output_dim + temporal_dim)
            # Adjust final FC layer input dim
            self.fc = nn.Linear(lstm_output_dim + temporal_dim, output_dim)
            self.dropout = nn.Dropout(dropout)

        def forward(self, x, temporal_features): # Expects both inputs
            # x shape: (batch_size, seq_length, input_dim)
            # temporal_features shape: (batch_size, temporal_dim)
            lstm_out, _ = self.lstm(x)
            lstm_out_norm = self.layer_norm_lstm(lstm_out)
            attn_intermed = torch.tanh(self.attention_fc1(lstm_out_norm))
            attn_scores = self.attention_fc2(attn_intermed)
            attn_weights = F.softmax(attn_scores, dim=1)
            context = torch.sum(attn_weights * lstm_out_norm, dim=1)

            # --- Concatenate temporal features ---
            if temporal_features.dim() == 1:
                 temporal_features = temporal_features.unsqueeze(0)
            if temporal_features.shape[0] != context.shape[0]:
                 # Handle potential batch size mismatch (e.g., last batch)
                 if temporal_features.shape[0] == 1 and x.shape[0] > 1:
                     temporal_features = temporal_features.repeat(context.shape[0], 1)
                 else:
                     raise ValueError(f"Batch size mismatch: context ({context.shape[0]}) vs temporal_features ({temporal_features.shape[0]})")
            if temporal_features.shape[1] != self.temporal_dim:
                 raise ValueError(f"Temporal feature dimension mismatch: expected {self.temporal_dim}, got {temporal_features.shape[1]}")

            combined_context = torch.cat((context, temporal_features), dim=1)
            context_norm = self.layer_norm_context(combined_context)
            out = self.dropout(context_norm)
            out = self.fc(out)
            return out
    # === End of NEW LSTM Model Class Definition ===


    # --- Instantiate the NEW DistillLagLSTMModel ---
    distill_lstm_input_dim = X_train_lstm_tensor.shape[2] # Input dim is 1 (lag value)
    distill_hidden_dim = 320 # Increased hidden dim
    distill_num_layers = 3   # Increased layers
    distill_dropout = 0.1
    distill_temporal_dim = num_temporal_features # Set the temporal dimension

    # Use the newly defined class
    distilled_lstm_model = DistillLagLSTMModel(
        input_dim=distill_lstm_input_dim,
        hidden_dim=distill_hidden_dim,
        output_dim=lstm_output_dim, # Assuming lstm_output_dim is defined (usually 1)
        num_layers=distill_num_layers,
        dropout=distill_dropout,
        attention_dim=lstm_attention_dim, # Assuming lstm_attention_dim is defined
        temporal_dim=distill_temporal_dim
    ).to(device)

    print("\nDistilled LSTM Model Architecture (Separate Class):")
    print(distilled_lstm_model)

    # --- Define Optimizer, Criterion, and Scheduler ---
    def combined_distillation_loss(student_output, teacher_target, gt_target, alpha=0.7):
        loss_teacher = F.mse_loss(student_output, teacher_target)
        loss_gt = F.mse_loss(student_output, gt_target)
        return alpha * loss_teacher + (1 - alpha) * loss_gt

    distill_criterion = combined_distillation_loss
    distill_optimizer = torch.optim.AdamW(distilled_lstm_model.parameters(), lr=0.001, weight_decay=1e-5)
    distill_scheduler = ReduceLROnPlateau(distill_optimizer, 'min', factor=0.2, patience=10, verbose=True)

    # --- Create DataLoaders (Input: Lags + Temporal, Target: Teacher + GT) ---
    distill_train_dataset = TensorDataset(X_train_lstm_tensor, train_temporal_tensor, y_train_teacher_target, y_train_gt_target)
    distill_train_loader = DataLoader(distill_train_dataset, batch_size=64, shuffle=True)
    distill_val_dataset = TensorDataset(X_val_lstm_tensor, val_temporal_tensor, y_val_teacher_target, y_val_gt_target)
    distill_val_loader = DataLoader(distill_val_dataset, batch_size=128, shuffle=False)

    # --- Training Loop ---
    distill_epochs = 300
    distill_patience = 40
    distill_patience_counter = 0
    best_distill_val_loss = float('inf')
    distill_train_losses = []
    distill_val_losses = []
    distill_model_save_path = "best_distilled_lstm_separate_model.pt" # New save path

    print("\nStarting Enhanced Distillation Training (Separate Model)...")
    for epoch in range(distill_epochs):
        distilled_lstm_model.train()
        epoch_loss = 0
        for batch_lags, batch_temporal, batch_y_teacher, batch_y_gt in distill_train_loader:
            distill_optimizer.zero_grad()
            # Pass both lags and temporal features to the forward method
            outputs = distilled_lstm_model(batch_lags, batch_temporal)
            loss = distill_criterion(outputs, batch_y_teacher, batch_y_gt, alpha=0.7)
            loss.backward()
            distill_optimizer.step()
            epoch_loss += loss.item()

        avg_train_loss = epoch_loss / len(distill_train_loader)
        distill_train_losses.append(avg_train_loss)

        # Validation
        distilled_lstm_model.eval()
        epoch_val_loss = 0
        with torch.no_grad():
            for batch_lags_val, batch_temporal_val, batch_y_teacher_val, batch_y_gt_val in distill_val_loader:
                val_outputs = distilled_lstm_model(batch_lags_val, batch_temporal_val)
                loss = distill_criterion(val_outputs, batch_y_teacher_val, batch_y_gt_val, alpha=0.7)
                epoch_val_loss += loss.item()

        avg_val_loss = epoch_val_loss / len(distill_val_loader)
        distill_val_losses.append(avg_val_loss)
        distill_scheduler.step(avg_val_loss)

        print(f"Epoch {epoch+1}/{distill_epochs}, Distill Train Loss: {avg_train_loss:.6f}, Distill Val Loss: {avg_val_loss:.6f}, LR: {distill_optimizer.param_groups[0]['lr']:.6f}")

        # Early stopping
        if avg_val_loss < best_distill_val_loss:
            best_distill_val_loss = avg_val_loss
            torch.save(distilled_lstm_model.state_dict(), distill_model_save_path)
            distill_patience_counter = 0
            print(f"  -> New best distillation val loss: {best_distill_val_loss:.6f}. Saving model.")
        else:
            distill_patience_counter += 1
            if distill_patience_counter >= distill_patience:
                print(f"Early stopping distillation training at epoch {epoch+1}")
                break

    # --- Load Best Distilled Model ---
    if os.path.exists(distill_model_save_path):
        # Re-instantiate the SEPARATE model class before loading
        distilled_lstm_model = DistillLagLSTMModel( # Use the separate class here too
            input_dim=distill_lstm_input_dim, hidden_dim=distill_hidden_dim,
            output_dim=lstm_output_dim, num_layers=distill_num_layers,
            dropout=distill_dropout, attention_dim=lstm_attention_dim,
            temporal_dim=distill_temporal_dim
        ).to(device)
        distilled_lstm_model.load_state_dict(torch.load(distill_model_save_path))
        print(f"\nBest SEPARATE distilled LSTM model loaded from '{distill_model_save_path}'.")
    else:
        print(f"\nWarning: '{distill_model_save_path}' not found. Using the model from the last epoch.")

    # --- Plot Distillation Losses ---
    if distill_train_losses and distill_val_losses:
        plt.figure(figsize=(10, 4))
        plt.plot(distill_train_losses, label='Distill Train Loss')
        plt.plot(distill_val_losses, label='Distill Val Loss')
        plt.title('Enhanced Distilled LSTM (Separate Model) Training History')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)
        plt.show()
    else:
        print("Skipping loss plot as training did not complete successfully.")
```

```python
# --- Evaluate Distilled Model on ORIGINAL Test Data ---
print("\nEvaluating ENHANCED distilled LSTM model on original test targets...")

# --- Define a NEW prediction function for the distilled model ---
def predict_distilled_in_batches(model, X_tensor_lags, temporal_tensor, batch_size=128):
    """Generates predictions in batches for a model expecting lags and temporal features."""
    model.eval()
    device = next(model.parameters()).device
    all_preds_scaled = []

    # Ensure tensors have the same length
    if len(X_tensor_lags) != len(temporal_tensor):
         raise ValueError("Length mismatch between X_tensor_lags and temporal_tensor in predict_distilled_in_batches")

    # Create dataset with both tensors
    dataset = TensorDataset(X_tensor_lags, temporal_tensor)
    dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=False)

    with torch.no_grad():
        for batch_lags, batch_temporal in dataloader: # Unpack both
            batch_lags = batch_lags.to(device)
            batch_temporal = batch_temporal.to(device)
            # Pass both to the model's forward method
            outputs = model(batch_lags, batch_temporal).squeeze()

            # Handle case where batch size is 1, squeeze might remove the batch dim entirely
            if outputs.dim() == 0:
                outputs = outputs.unsqueeze(0)
            all_preds_scaled.append(outputs.cpu().numpy())

    # Concatenate results from all batches
    if not all_preds_scaled: # Handle empty prediction case
        return np.array([])
    predictions_scaled = np.concatenate(all_preds_scaled)
    return predictions_scaled.flatten() # Ensure 1D array
# --- End of new prediction function ---


# Ensure the distilled model exists and is loaded
if 'distilled_lstm_model' not in locals():
    print("Error: 'distilled_lstm_model' not found. Run the distillation training cell first.")
    distilled_test_pred_scaled = np.array([]) # Assign empty array
else:
    distilled_lstm_model.eval()
    with torch.no_grad():
        # Predict using the LAGS and TEMPORAL test input tensors
        # Ensure X_test_lstm_tensor and test_temporal_tensor exist from cell 663da4ae
        if 'X_test_lstm_tensor' in locals() and 'test_temporal_tensor' in locals():
            # --- MODIFIED: Use the NEW prediction function ---
            try:
                distilled_test_pred_scaled = predict_distilled_in_batches( # Call the new function
                    distilled_lstm_model,
                    X_test_lstm_tensor,      # Lags
                    test_temporal_tensor,    # Temporal features
                    batch_size=128
                )
            except Exception as e:
                 print(f"An unexpected error occurred during prediction using predict_distilled_in_batches: {e}")
                 import traceback
                 traceback.print_exc() # Print detailed traceback
                 distilled_test_pred_scaled = np.array([])

        else:
            print("Error: Test input tensors ('X_test_lstm_tensor', 'test_temporal_tensor') not found. Cannot evaluate.")
            distilled_test_pred_scaled = np.array([]) # Assign empty array to avoid further errors
```

```python
# Proceed only if predictions were generated
if distilled_test_pred_scaled.size > 0:
    # Inverse transform predictions
    distilled_test_pred_orig = target_scaler.inverse_transform(distilled_test_pred_scaled.reshape(-1, 1)).flatten()
    distilled_test_pred_orig = np.clip(distilled_test_pred_orig, 0, None)

    # Calculate metrics against the ORIGINAL test targets
    # Ensure y_test_orig_final exists and has the same length
    if 'y_test_orig_final' in locals() and len(y_test_orig_final) == len(distilled_test_pred_orig):
        distilled_rmse = np.sqrt(mean_squared_error(y_test_orig_final, distilled_test_pred_orig))
        distilled_mae = mean_absolute_error(y_test_orig_final, distilled_test_pred_orig)

        print(f"\nDistilled LSTM (Temporal) - Test RMSE: {distilled_rmse:.2f}")
        print(f"Distilled LSTM (Temporal) - Test MAE: {distilled_mae:.2f}")

        # --- Optional: Add to Results Comparison ---
        print("\n--- Comparison Including Distilled LSTM (Temporal) ---") # Modified print
        # Ensure previous metrics exist
        if 'lstm_test_rmse' in locals() and 'test_rmse_xgb' in locals() and 'test_rmse_nn' in locals():
             comparison_data = {
                 'Model': ['LSTM Only', 'LSTM + XGBoost', 'LSTM + NN', 'Distilled LSTM (Temporal)'], # Modified label
                 'RMSE': [lstm_test_rmse, test_rmse_xgb, test_rmse_nn, distilled_rmse],
                 'MAE': [lstm_test_mae, test_mae_xgb, test_mae_nn, distilled_mae]
             }
             comparison_df = pd.DataFrame(comparison_data)
             print(comparison_df.sort_values('RMSE'))
        else:
            print("Warning: Could not generate full comparison table, previous model metrics missing.")


        # --- Optional: Plot Comparison ---
        plt.figure(figsize=(15, 7))
        # Define indices for the last 24 hours, ensuring start index is not negative
        start_index = max(0, len(y_test_orig_final) - 24)
        end_index = len(y_test_orig_final)
        plot_indices = range(start_index, end_index)
        plt.plot(plot_indices, y_test_orig_final[start_index:end_index], label='Actual', marker='.', linestyle='-', alpha=0.7)
        if 'lstm_test_pred_orig' in locals() and len(lstm_test_pred_orig) == len(y_test_orig_final):
            plt.plot(plot_indices, lstm_test_pred_orig[start_index:end_index], label=f'LSTM Only (RMSE: {lstm_test_rmse:.2f})', marker='^', linestyle=':', alpha=0.8)
        if 'final_test_pred_xgb' in locals() and len(final_test_pred_xgb) == len(y_test_orig_final):
            plt.plot(plot_indices, final_test_pred_xgb[start_index:end_index], label=f'LSTM + XGBoost (RMSE: {test_rmse_xgb:.2f})', marker='x', linestyle='--', alpha=0.9)
        if 'final_test_pred_nn' in locals() and len(final_test_pred_nn) == len(y_test_orig_final):
            plt.plot(plot_indices, final_test_pred_nn[start_index:end_index], label=f'LSTM + NN (RMSE: {test_rmse_nn:.2f})', marker='s', linestyle='-.', alpha=0.9)
        # Plot distilled LSTM results
        plt.plot(plot_indices, distilled_test_pred_orig[start_index:end_index], label=f'Distilled LSTM (Temporal) (RMSE: {distilled_rmse:.2f})', marker='d', linestyle='-', alpha=0.8) # Modified label

        plt.title('Comparison of Models (Including Distilled LSTM with Temporal) on Test Set') # Modified title
        plt.xlabel('Time Step')
        plt.ylabel('Vehicle Count')
        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print("Error: Cannot calculate metrics or plot. Mismatch between prediction length and target length or 'y_test_orig_final' missing.")

# Add a final else for the outer check
else:
    print("Skipping evaluation and plotting for Distilled LSTM due to prediction generation error.")
```
